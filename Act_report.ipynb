# Bloc Montrant les Visualisation du Projet WeRateDogs

Dans ce présent projet, il a été question d'exploiter les données de Twitter de WeRateDogs pour créer des analyses et des visualisations intéressantes et fiables. 

Nous avions donc dans le cadre de ce projet après rassemblage, nettoyage, stockage et analyse, effectuer les visualisations suivantes :

### 1. Diagramme circulaire montrant le pourcentage des espèces de chiens représentés

```python
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
 
#plt.xlabel("X pixel scaling")
#plt.ylabel("Y pixels scaling")
 
image = mpimg.imread("diagramme_circulaire.png")
plt.figure(figsize = (8,8))
plt.imshow(image)
plt.show()
```

- Pupper est l'espèce de chien la plus représentée (68.4%).
- Floofer est l'espèce de chien la moins représentée (0.9%).

### 2. Diagramme en barre montant les espèces de chiens les plus représentés

### 3. Histogramme montrant la distribution des colonnes numériques

### 6. L'impact du temps sur les retweets et les favoris compte.

```python
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
 
#plt.title("Histogramme montrant la distribution des colonnes numériques")
#plt.xlabel("X pixel scaling")
#plt.ylabel("Y pixels scaling")
 
image = mpimg.imread("imapct_temps_retweet.png")
plt.figure(figsize = (8,8))
plt.imshow(image)
plt.show()
```
